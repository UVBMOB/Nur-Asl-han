```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

sample = np.random.poisson(20.,size=[10000])      # dagilimdan ornek al
(x,y) = np.unique(sample,return_counts=True)  
y = np.array(y).astype(np.float32)
y /= np.sum(y)                                    # normallestir

p20 = poisson(20)
pmf20 = np.array([p20.pmf(i) for i in x])

plt.plot(x,pmf20)                                # olmasi gereken dagilim
plt.plot(x,y)                                    # rastgele orneklem
plt.title("Poisson 20'dan Orneklem")
plt.legend(["Ideal", "Orneklem"])
plt.savefig("sample.png")
```

```python
# dagilimlari olusturalim

p2 = poisson(2)
p5 = poisson(5)
p11= poisson(11)

n = 30
x = np.linspace(0,n,n+1) 
pmf2 = np.array([p2.pmf(i) for i in x])
pmf5 = np.array([p5.pmf(i) for i in x])
pmf11 = np.array([p11.pmf(i) for i in x])

plt.plot(x, pmf2, 'r')
plt.scatter(x,pmf2,c='r')
plt.plot(x, pmf5, 'g')
plt.scatter(x, pmf5,c='g')
plt.plot(x, pmf11, 'b')
plt.scatter(x, pmf11,c='b')
plt.title("Poisson Olasilik Kutle Fonksiyonlari")
plt.legend(["Poisson(2)","Poisson(5)","Poisson(11)"])
plt.show()
```

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

# dagilimlari olusturalim

p2 = poisson(2)
p5 = poisson(5)
p11 = poisson(11)

n = 30
x = np.linspace(0,n,n+1) 
cdf2 = np.array([p2.cdf(i) for i in x])
cdf5 = np.array([p5.cdf(i) for i in x])
cdf11 = np.array([p11.cdf(i) for i in x])

plt.plot(x, cdf2, 'r')
plt.scatter(x,cdf2,c='r')
plt.plot(x, cdf5, 'g')
plt.scatter(x, cdf5,c='g')
plt.plot(x, cdf11, 'b')
plt.scatter(x, cdf11,c='b')
plt.title("Poisson Kumulatif Olasilik Kutle Fonksiyonlari")
plt.legend(["Poisson(2)","Poisson(5)","Poisson(11)"])
plt.xticks(x)
plt.savefig("pmf.png")
```
